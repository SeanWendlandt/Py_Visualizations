```python
# Part 1: Using the built-in Seaborn dataset mpg, provide the heatmap of the correlation of all the numeric columns and provide the pairplot of the same.

mpgdf= sns.load_dataset("mpg")
mpgdf = mpgdf.drop(columns=["origin", "name"])

#Heatmap
sns.heatmap(mpgdf.corr(), cmap="Greens")

# scaling data for better visualization
#scaler = MinMaxScaler(feature_range=(20, 500))
#mpgdf[["weight"]] = scaler.fit_transform(mpgdf[["weight"]])


#Pairplot
sns.pairplot(mpgdf, height=1)
```

```python
#Part 2: Using the built-in Seaborn dataset diamonds, establish a FacetGrid based on ‘cut’ and ‘color’.  Eliminate colors ‘D’ and ‘E’ as well as the cut ‘Fair’.  Within that grid, plot the scatterplot for  ‘price’ vs. ‘carat’.

diamondsdf = sns.load_dataset("diamonds")
diamondsdf.color, diamondsdf.cut = diamondsdf.color.cat.remove_categories(["D", "E"]), diamondsdf.cut.cat.remove_categories("Fair")
diamondsgrid = sns.FacetGrid(diamondsdf.dropna(), row="cut", col="color")
diamondsgrid.map(sns.scatterplot, "carat", "price")
```

```python
#Part 3: Using the built-in Seaborn dataset car_crashes, prepare plots with the scatter gram with the linear model for both the total vs. speeding and the total vs. alcohol.
crashesdf = sns.load_dataset("car_crashes")
#print(crashesdf)
speeding = sns.lmplot(x="total", y="speeding", data=crashesdf)
alcohol = sns.lmplot(x="total", y="alcohol", data=crashesdf)
```

```python
#Part 4: Using the built-in Seaborn dataset iris, provide a plot with four subplots wherein the distribution of each of the numeric columns is presented as a set of boxplots, one for each ‘species’.
irisdf = sns.load_dataset("iris")
fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y= "petal_width", x= "species", data=irisdf, orient='v', palette="Set1", ax=axes[0, 0])
sns.boxplot( y= "petal_length", x= "species", data=irisdf, orient='v', palette="RdPu_r", ax=axes[0, 1])
sns.boxplot( y= "sepal_length", x= "species", data=irisdf, orient='v', palette="bone_r", ax=axes[1, 0])
sns.boxplot( y= "sepal_width", x= "species", data=irisdf, orient='v', palette="Spectral", ax=axes[1, 1])
plt.show()
```

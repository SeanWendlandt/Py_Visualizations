```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights=sns.load_dataset('flights')
iris=   sns.load_dataset('iris')
tips=   sns.load_dataset('tips')
titanic=sns.load_dataset('titanic')

print(flights.head())
print(iris.head())
print(tips.head())
print(titanic.head())
```

```python
sns.set()
sns.set_style('whitegrid')
sns.lmplot(x='tip', y='total_bill', data=tips, aspect=2)
plt.show()


# g3=sns.lmplot(x='tip', y='total_bill', data=tips, aspect=2)
#plt.show(g3)
```

```python
sns.set_style('darkgrid')
sns.FacetGrid(titanic, col='survived', row='sex')\
.map(plt.hist,'age')
plt.show()
```

```python
sns.FacetGrid(iris, col='species')\
.map(plt.scatter,'sepal_length','sepal_width')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='tip', y='total_bill',
  data=tips, aspect=2, fit_reg=False)
plt.show()
```

```python
sns.regplot(x=iris['sepal_length'],
  y=iris['sepal_width'], fit_reg=False)
plt.show()
```

```python
sns.lmplot(data=iris, x='sepal_length',
  y='sepal_width', hue='species',
  fit_reg=False)
plt.show()
```

```python
plt.scatter(x=iris['sepal_length'],
  y=iris['sepal_width'])
plt.show()
```

```python
sns.set_style('whitegrid')
sns.PairGrid(iris).map(plt.scatter)
plt.show()
```

```python
sns.pairplot(iris)  # Better than above
plt.show()
```

```python
names = ['anscombe', 'attention',
 'car_crashes', 'diamonds', 'dots', 'exercise', 'flights', 'fmri', 'gammas', 'iris', 'mpg', 'planets', 'tips', 'titanic']
print(names)

for name in names:
    print('\n\n', name)
    df=sns.load_dataset(name)
    print(df.head())
    print(df.info())
    print(df.describe())
    print(df.corr())
    sns.heatmap(df.corr(), vmin=-1, annot=True, cmap='coolwarm')
    plt.show()
    sns.pairplot(df)
    plt.show()
```

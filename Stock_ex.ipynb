```python
# Part 1
# First 6 Atomic Weights for Pie Chart Values
Weights = [1, 4, 7, 9, 11, 12]

# Pie Chart 1 

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
exploded = [0, 0, 0, 0, 0, 0.3]
ax.pie(Weights, autopct='%1.2f%%',explode=exploded) 
plt.show()

# Pie Chart 2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
exploded = [0, 0, 0, 0.35, 0, 0]
ax.pie(Weights, labels=Weights, explode=exploded) 

plt.show()
```

```python
# Part 2
# Reading in csv to dataframe
df_2=pd.read_csv("py_ide2.csv")
df_2.head()
# Horizontal Bar Chart
df_2.plot.barh(x='IDE', y='Adoption')
plt.xticks(rotation=45, fontsize='small')

# Vertical Bar Chart
df_2.plot.bar(x='IDE', y='Adoption')
plt.xticks(rotation=45, fontsize='small')
plt.show()
```

```python
# Part 3
# List of 8 strings representing 1st through 8th of a month
day_l = ['1-1-22', '1-2-22', '1-3-22', '1-4-22', '1-5-22', '1-6-22', '1-7-22', '1-8-22']

# 8 uniform samples from ranging between 100 and 200
a_ran = np.random.uniform(100,200,8)

# making a df
df3 = pd.DataFrame({'Days': day_l, 'Dist': a_ran})
# making days into datetime
df3['Days'] = pd.to_datetime(df3['Days'])
# setting days to index
df3 = df3.set_index('Days')
#print(df3)

# Generating plot
fig, axes = plt.subplots(nrows=2, ncols=1)

pl1=df3.plot(ax=axes[0])
pl1.set_ylim(bottom=100, top=200)
fig.tight_layout()

pl2=df3.plot.bar(ax=axes[1])
pl2.set_ylim(bottom=0, top=200)
plt.xticks(rotation=45, fontsize='small')
fig.tight_layout()
```

```python
# Pulling data from yahoo finance for GameStop
GMEdata = yf.download("GME", start='2022-01-01', end='2022-02-01')
#print(GMEdata)
close = GMEdata['Close']
volume = GMEdata['Volume']


# Generating plot
fig, axes = plt.subplots(nrows=2, ncols=1)

pl1=close.plot(ax=axes[0])
axes[0].set(xlabel="Date", ylabel="Price")
fig.tight_layout()

pl2=volume.plot(ax=axes[1])
plt.xticks(rotation=45, fontsize='small')
axes[1].set(xlabel="Date", ylabel="Volume")
fig.tight_layout()
```
